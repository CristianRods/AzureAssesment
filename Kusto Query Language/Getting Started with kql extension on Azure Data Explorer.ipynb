## Query data

```kusto
StormEvents
| where isnotempty(EndLocation) 
| summarize event_count=count() by EndLocation
| order by event_count
```

```kusto
StormEvents
| summarize event_count=count(), mid = avg(BeginLat) by State
| sort by mid
| where event_count > 1800
| project State, event_count
```

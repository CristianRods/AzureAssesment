# SQL Assessment - Microsoft's Best Practices Checker

## Abstract

<span style="font-size: 14px;">Do you want to get the current 232 suggested best practices from the SQL Server Team and evaluate your environments? These best practices are available for all instances, from your Data Center to the cloud.</span>

<span style="font-size: 14px;">In this all demo session (Yes, no slides), I will show you both the "what" and "how" to use the SQL assessment API. You will learn if a setting is changed after your initial check. How to customize your checks and create your own rules that fit your environments? The demo will use SQL assessment API, PowerShell cmdlets, and the new 'SQL Assessment' extension in Azure Data Studio to export assessment results in a SQL table and create an HTML report.</span>

At the end of this session, you will be equipped with the toolset to set up and run your own assessment, create customized rules, and determine your compliance score against suggested best practices.

### Why Would You Listen to This Talk?

- Is my SQL Server, SQL Server on Azure VM, Azure SQL Managed Instance configured with Microsoft recommended best practice?
    - Best practice rules are suggested by SQL Server Team.
    - Users can tune the default rules.
    - Create your own.
    - Bring rules created by third party. <mark>_(Do not use rulesets that come from untrusted sources until you thoroughly review them to make sure they are safe)_</mark>
    - Users can disable certain rules.
